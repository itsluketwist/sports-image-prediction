````python
# ONLY NEEDS RUNNING ONCE. BE CAREFUL WHEN EDITING.
# This cell simply loads the model, and sets up the prediction method.

import torch
from src.predict import make_prediction, load_sports_image
from IPython.display import Image, display

MODEL_PATH = "output/sports_mod_tuned.pth"  # change this to use a different model

loaded_model = torch.load(MODEL_PATH)

def what_sport(file: str) -> str:
    _image, _label = load_sports_image(path=file)
    [_prediction] = make_prediction(
        model=loaded_model,
        data=[(_image, _label)],
        display=False,
    )
    print(f"Actual label: {_label}")
    print(f"Prediction: {_prediction} {'‚úîÔ∏è' if _label == _prediction else '‚úñÔ∏è'}")
    display(Image(filename=file))
````

````python
PREDICT_THIS_FILE = "sample/baseball.png"

what_sport(file=PREDICT_THIS_FILE)
````

````python
PREDICT_THIS_FILE = "sample/basketball.png"

what_sport(file=PREDICT_THIS_FILE)
````
